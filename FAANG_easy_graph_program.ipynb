```
#Importo librerie per costruzione grafico e dati
from datetime import datetime
import numpy as np
import pandas as pd
import pandas_datareader as web
import matplotlib.pyplot as plt 
plt.style.use('fivethirtyeight')
```

```
#Tracciamo FAANG
stockSymbols = ["FB", "AMZN", "AAPL", "NFLX", "GOOG"]
```

```
#Data inzio tracciamento
stockStartDate = '2014-01-01'
```

```
#Data di oggi nel formato anno-mese-giorno
today = datetime.today().strftime('%Y-%m-%d')

print(today)
```

```
#funzione che prende i dati dal mio portfolio
def MyPortfolio (stocks = stockSymbols, start = stockStartDate, end = today, col = 'Adj Close'):
  data = web.DataReader(stocks, data_source = 'yahoo', start = start, end = end)[col]
  return data 
```

```
my_stocks = MyPortfolio(stockSymbols)
my_stocks
```

```
#Creare una funzione per visualizzare grafico
def Graph (stocks = stockSymbols, start = stockStartDate, end = today, col ='Adj Close'):
  
  title = 'Price' +col+ 'History'
  
  my_stocks = MyPortfolio (stocks = stocks, start = start, end = end, col = col)

  plt.figure(figsize = (12.1, 4.5))

  for x in my_stocks.columns.values:
    plt.plot(my_stocks[x], label = x)
  
  #Misure
  plt.title(title)
  plt.xlabel('Date', fontsize = 16)
  plt.ylabel(col + 'Price USD($)', fontsize = 16)
  plt.legend(my_stocks.columns.values, loc = 'upper left')
  plt.show() #Permette di visualizzare il grafico
```

```
 #Grafico finale
 Graph(stockSymbols)
```
